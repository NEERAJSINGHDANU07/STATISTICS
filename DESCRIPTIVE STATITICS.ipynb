## **RANGE**
- RANGE IS THE DIFFERNCE BETWEEN THE LARGEST AND SMALLEST DATA POINTS IN THE SET.
- RANGE = MAXIMUN VALUE - MINIMUM VALUE
- FOR OUR PRIVIOUS EXAMPLE
    - MEAN OF THE PLAYER 1 = 15
    - MEAN OF THE PLAYER 2 = 15
 - RANGE OF THE PLAYER 1 = 25-5 = 20
 - RANGE OF THE PLAYER 2 = 20 - 10 = 10
      - THUS THE PLAYER 2 IS MORE CONSISTANCE THAN PLAYER 1 AND HAS A MORE RELIABLE AVERAGE

## **PROBLEM WITH THE RANGE**
- **SENSITIVE TO OUTLIERES** : THE RANGE IS HIGHLY SENSTIVE TO EXTREME VALUES (OUTLIERS) BECAUSE IT DEPEND ONLY ON THE MAXIMUN AND MINIMUM
  VALUES. A SINGLE EXTREME VALUE CAN SIGNIFICANTLY DISORT THE MEASURE
- EXAMPLE : CONSIDER THE FOLLOWING DATASET: {10,12,15,18,20,100}. THE RANGE IS 100 - 10 = 90. HOWEVER THE MAJORITY OF THE VALUE ARE WITHIN A SMALL RNAGE AND THE LARGE OUTLIER 100  GREAT INFLUENCES THE RESULT

## **INTERQUARTILE RANGE**
- IQR ek aisi value hoti hai jo batati hai ki data ke beech ke 50% values (na bahut chhoti, na bahut badi)kitni range mein aati hain.
- INTERQUARTILE RANGE IS THE DIFFERENCE BETWEEEN THE FIRST(Q1) AND THE THIRD (Q3)
- IR DESCRIBE THE MIDDLE 50 % OF THE OBSERVATRIONS. IF THE MIDDLE 50 % OF THE DATA POINTS AER SPACED WIDE APART ,THEIR INTERQUATILE RANGE WILL BE LARGE ***[ IQR = Q3-Q1]***
- THE IQR IS PRIMARLIY USED FOR DETECTING OUTLIERS . VALUES BELOW  ***Q1 - 1.5 IQR*** AND ABOVE ***Q3 + 1.5 IQR*** ARE CONSIDERED AS OUTLIERS

## **Z SCORE**
- ALSO CALLED THE STANDARD SCORE
- IT GIVES AN IDEA OF HOW FAR FROM THE MEAN A DATA POINT IS

- WHEN USING Z SCORES IT IS ASSUMED  THAT THE DATA IS NORMALLY DISTRUBED

## KURTOSIS 
- KURTOSIS IS A MEASURE OF THE "TAILEDNESS" OF A DISTRUBUTION.
- DATA SETS WITH HIGH KURTOSIS TEND TO HAVE HEAVY TAILS OR OUTLIERS . DATA SETS WITH LOW KURTOSIS TEND TO HAVE LIGHT TAILS, OR  LACK OF OTLIERS
- KURTOSIS RANGES FROM 1 TO INIFINITY
- A NORMAL DISTRIBUTION HAS A KURTOSIS OF 3
- KURTOSIS IS SOMETIMES SEEN AS THE "PEAKEDNESS" OF A DISTRIBUTION WHICH IS INCORRECT.
