## **TYPES OF ESTIMATES**
-  **POINT ESTIMATE:** A POINT ESTIMATE IS A SINGLE VALUE THAT SERVES AS AN APPROXIMATION OF A POPULATION PARAMETER. COMMON POINT ESTIMATES INCLUDE THE SAMPLE MEAN, MEDIAN, PROPORTION, AND STANDARD DEVIATION, WHICH ESTIMATE THE CORRESPONDING POPULATION PARAMETERS.
- **CONFIDENCE INTERVAL:** A CONFIDENCE INTERVAL, ON THE OTHER HAND, PROVIDES A RANGE OF VALUES WITHIN WHICH THE TRUE POPULATION PARAMETER IS LIKELY TO FALL, WITH A CERTAIN LEVEL OF CONFIDENCE.
- EG - CONTINUING WITH THE HEIGHT ESTIMATION, SUPPOSE THE AVERAGE HEIGHT OF YOUR SAMPLE IS 175 CM, AND YOU CALCULATE A 95% CONFIDENCE INTERVAL FROM 170 CM TO 180 CM. THIS INTERVAL SUGGESTS THAT YOU ARE 95% CONFIDENT THAT THE TRUE AVERAGE HEIGHT OF ALL ADULT MALES IN THE CITY IS BETWEEN 170 CM AND 180 CM.
-  WE BUILD THE CONFIDENCE INTERVAL AROUND THE POINT ESTIMATE.
-  A CONFIDENCE INTERVAL IS CONSTRUCTED ON THE SAMPLING DISTRIBUTION OF A STATISTIC.
- (1-α) IS THE LEVEL OF CONFIDENCE. WE ARE (1-α) * 100% CONFIDENT THAT THE POPULATION PARAMETER WILL FALL IN THE SPECIFIED INTERVAL. COMMON ALPHAS ARE: 0.01, 0.05, 0.1.



## **CONFIDENCE INTERVALS**
- GENERAL FORMULA TO COMPUTE CONFIDENCE INTERVAL:
[POINT ESTIMATE - MARGIN OF ERROR , POINT ESTIMATE + MARGIN OF ERROR]

- THE MARGIN OF ERROR CAN BE CALCULATED EITHER USING Z DISTRUBUTION (STANDARD NORMAL DISTRUBUTION ) OR THE STUDENT'S T DISTRUBUTION

## EXAMPLE

LET'S SAY WE WANT TO ESTIMATE THE AVERAGE HEIGHT OF ADULT MEN IN A CITY. WE TAKE A RANDOM SAMPLE OF 100 MEN AND FIND THEIR AVERAGE HEIGHT TO BE 175 CM WITH A STANDARD DEVIATION OF 10 CM. WE CALCULATE A 95% CONFIDENCE INTERVAL FOR THE TRUE AVERAGE HEIGHT.1. CALCULATE THE STANDARD ERROR (SE):$$SE = \frac{\text{STANDARD DEVIATION}}{\sqrt{\text{SAMPLE SIZE}}} = \frac{10}{\sqrt{100}} = 1 \text{ CM}$$2. DETERMINE THE CRITICAL VALUE FOR A 95% CONFIDENCE LEVEL:FOR A 95% CONFIDENCE LEVEL, THE VALUE (Z-SCORE) IS APPROXIMATELY 1.96.3. CALCULATE THE 95% CONFIDENCE INTERVAL (CI):$$\text{CI} = \text{SAMPLE MEAN} \pm \text{MARGIN OF ERROR}$$$$\text{CI} = 175 \pm 1.96$$LOWER BOUND: $175 - 1.96 = 173.04 \text{ CM}$UPPER BOUND: $175 + 1.96 = 176.96 \text{ CM}$
CONCLUSION: WE ARE 95% CONFIDENT THAT THE TRUE AVERAGE HEIGHT OF ADULT MEN IN THE CITY LIES BETWEEN 173.04 CM AND 176.96 CM.

 ### WHAT IS HYPOTHESIS TEST ?
- A HYPOTHESIS TEST HELPS US TO UNDERSTAND WHETHER RANDOM CHANCE MIGHT BE RESPONSIBLE FOR AN OBSERVED EFFECT.

- EG–
- PATIENTS TAKING A NEW MEDICATION SEEM TO RECOVER FASTER THAN THOSE NOT TAKING IT. IT MAY BE POSSIBLE THAT THE RECOVERY MAY BE DUE TO THE PLACEBO EFFECT, AND NOT DUE TO THE EFFICACY OF THE MEDICATION.
- HYPOTHESIS: PRICE B PRODUCES MORE PROFIT THAN PRICE A. THIS MAY BE DUE TO SOME OTHER REASON, MAYBE DUE TO SOME FESTIVAL.
- STATISTICAL HYPOTHESIS TESTING PROTECTS US FROM UNDERESTIMATING THE SCOPE OF NATURAL RANDOM BEHAVIOR.


## ***NULL (H₀) & ALTERNATE (H₁) HYPOTHESES***

- IN DEVELOPING A HYPOTHESIS, WE NEED TO SETUP TWO STATEMENTS, NULL & ALTERNATE HYPOTHESIS.
- **NULL HYPOTHESIS:** THE HYPOTHESIS THAT CHANCE IS RESPONSIBLE FOR ANY EFFECT OR CHANGE.
- **ALTERNATE HYPOTHESIS:** THIS IS THE COUNTER HYPOTHESIS WHICH WE HOPE TO PROVE. THE ALTERNATE HYPOTHESIS STATES THAT THE EFFECT IS NOT DUE TO RANDOM CHANCE BUT DUE TO A TRUE DIFFERENCE BETWEEN A AND B.
- EG–
- **NULL HYPOTHESIS (H₀):** THE NEW DRUG HAS NO EFFECT ON BLOOD PRESSURE COMPARED TO THE CURRENT STANDARD TREATMENT.
- **ALTERNATIVE HYPOTHESIS (H₁):** THE NEW DRUG LOWERS BLOOD PRESSURE MORE EFFECTIVELY THAN THE CURRENT STANDARD TREATMENT.
- **NULL HYPOTHESIS (H₀):** CLASS SIZE HAS NO IMPACT ON STUDENT TEST SCORES.
- **ALTERNATIVE HYPOTHESIS (H₁):** STUDENTS IN SMALLER CLASSES ACHIEVE HIGHER TEST SCORES THAN THOSE IN LARGER CLASSES.

## **CORRELATION**
- **WHAT IS CORRELATIONS ?**
  - CORRELATIONIS A STATISTICAL MEASURE THAT DESCRIBS THE STRENGTH AND DIRECTIONS OF A RELATIONSHIP BETWEEN TWO VARIABLE . IT QUANTIFIES HOW CHANGES IN ONE VARIABLE ARE ASSOCIATED WITH CHANGES IN ANOTHER . THE CORRELATION COEFFOCIENT , USUALLY DENOTED (r) , RANGE FROM -1 TO 1
  -  r = 1 , PREFECT POSITIVE CORRELATIONS (AS ONE VARIABLE INCREASES), THE OTHER INCREASES PROPORTIONALLY
  -  r = -1, PREFECT POSITIVE CORRELATIONS (AS ONE VARIABLE INCREASES), THE OTHER DECREASE PROPORTIONALLY
  -  r = 0 , NO CORRELATIONS (NO LINEAR RELATIONSHIP BETWEEN THE VARIABLE )

## **CORRELATION AND CAUSATION**

THE PHRASE "CORRELATION IS NOT CAUSATION" MEANS THAT JUST BECAUSE TWO VARIABLES ARE CORRELATED (I.E., THEY MOVE TOGETHER IN SOME WAY), IT DOES NOT NECESSARILY MEAN THAT ONE VARIABLE CAUSES THE OTHER. CORRELATION INDICATES A RELATIONSHIP OR ASSOCIATION BETWEEN VARIABLES, BUT CAUSATION INDICATES THAT ONE VARIABLE DIRECTLY AFFECTS THE OTHER. HERE ARE A FEW EXAMPLES TO ILLUSTRATE THIS CONCEPT:

- EXAMPLE 1: ICE CREAM SALES AND DROWNING INCIDENTS

• CORRELATION: STUDIES MAY SHOW THAT ICE CREAM SALES AND DROWNING INCIDENTS BOTH INCREASE DURING THE SUMMER MONTHS.

• EXPLANATION: WHILE THERE IS A CORRELATION BETWEEN ICE CREAM SALES AND DROWNING INCIDENTS, BUYING ICE CREAM DOES NOT CAUSE DROWNING. THE ACTUAL CAUSE IS THE HOT WEATHER, WHICH LEADS TO MORE PEOPLE BUYING ICE CREAM AND MORE PEOPLE SWIMMING, THUS INCREASING THE RISK OF DROWNING.

- EXAMPLE 2: SHOE SIZE AND READING ABILITY IN CHILDREN

• CORRELATION: DATA MIGHT SHOW A POSITIVE CORRELATION BETWEEN SHOE SIZE AND READING ABILITY IN CHILDREN.

• EXPLANATION: LARGER SHOE SIZES DO NOT CAUSE BETTER READING ABILITY. THE UNDERLYING FACTOR IS AGE. OLDER CHILDREN TEND TO HAVE LARGER SHOE SIZES AND BETTER READING SKILLS BECAUSE THEY HAVE HAD MORE TIME TO DEVELOP THESE SKILLS.
